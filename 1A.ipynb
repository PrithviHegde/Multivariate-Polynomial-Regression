Here, we import the necessary libraries for assignment 1-A,

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

```

We plot our initial beta-distribution (with a,b both equal to 2).
<br><br>
We know that the initial (prior) beta-distribution is given by,

<br>


and our beta-distribution post updation (posterior), is given by,


```python
#initial conditions of the beta-distribution
arr = np.linspace(0,1,50)
al = 2
be = 2
y = beta.pdf(arr,al,be)
plt.title("Initial distribution without update (prior)")
plt.plot(arr,y)

plt.show()
```

Here, we know that in survey 1,<br>
40 people like and 10 people dislike the update- and therefore we update the parameters as such.



```python
#after the first survey, 40 liked and 10 disliked- since the distribution is conjugate, we update accordingly
al += 40
be += 10
y = beta.pdf(arr,al,be)
plt.title("Beta distribution after the first update (first posterior)")
plt.plot(arr,y)

plt.show()
```

For our likelihood function, we have <br>
<b>N = 50, and m = 40</b>, and therefore, as per the formula, we can calculate the likelihood as:
<br>


```python
coeff = np.math.factorial(50)/(np.math.factorial(10)*np.math.factorial(40))
y = (coeff)*pow(arr,40)*pow(1-arr,10)
plt.plot(arr,y)
plt.title("Likelihood function for N=50, m=40")

plt.show()
```

In our next survey, we find that 13 people liked and 17 people disliked the update. <br>Therefore, our old priors (the beta distribution with a = 42, b = 12) now get updated as 
<b> a = 55, and b = 29. </b>
(We plot the likelihoods later).
<br><br>
In the next survey, we find that 70 people liked and 30 people disliked the update. <br>
Our priors (a = 55, b = 29) get updated to become <b> (a = 125, and b= 59). </b>



```python
#in next survey, 13 liked and 17 disliked. We update accordingly.
#(in this case, the prior is the preceding distribution, as we have incorporated its' knowledge into the distribution)
al += 13
be += 17
y1 = beta.pdf(arr,al,be)
#in this survey, we find that 70 liked and 30 disliked, and update.
#(again, the preceding distribution is the prior)
al += 70
be += 30
y2 = beta.pdf(arr,al,be)

fig,ax = plt.subplots(1,2)

ax[0].set_title("after second update")
ax[0].plot(arr,y1)
ax[1].set_title("after third update")
ax[1].plot(arr,y2)

plt.show()
```

Now, we plot the likelihoods of the two updates:<br>
In the first update, we find that N = 30, and m = 13, and in the second, we find <br> 
that N = 100, and m = 70.



```python
coeff1 = np.math.factorial(30)/(np.math.factorial(17)*np.math.factorial(13))
y1 = (coeff1)*pow(arr,13)*pow(1-arr,17)
coeff2 = np.math.factorial(100)/(np.math.factorial(70)*np.math.factorial(30))
y2 = (coeff2)*pow(arr,70)*pow(1-arr,30)


fig,ax = plt.subplots(1,2)

ax[0].plot(arr,y1)
ax[0].set_title("Likelihood for N=30, m=17")
ax[1].plot(arr,y2)
ax[1].set_title("Likelihood for N=100, m=70")

plt.show()
```
